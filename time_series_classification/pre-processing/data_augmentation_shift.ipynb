```python
idx = data['idx']
# for i in idx:
#     print(idx[i])

temp = data.iloc[1]
# print(temp)

def divide_list(l,n):
    for i in range(0, len(l), n):
        yield l[i:i+n]

temp = list(temp)
del temp[0]

t = list(divide_list(temp,60))
len(t)

```

```python
list_60 = []

for i in range(len(idx)):
    temp = data.iloc[i]
    temp = list(temp)
    del temp[0]
    temp1 = list(divide_list(temp,60))
    list_60 = list_60 + temp1

print(len(list_60))
```
