# **Data Load**

```python
# System related and data input controls
import os
from glob import glob

# Data manipulation and visualization
import pandas as pd
pd.options.display.float_format = '{:,.2f}'.format
pd.options.display.max_rows = 100
pd.options.display.max_columns = 20
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
# Data_kimm_op - 유선진동 정상시계열
# Label_kimm_op - 유선진동 정상시계열 라벨
# Data_kimm_no - 유선진동 고장시계열
# Label_kimm_no - 유선진동 고장시계열 라벨
label = ['0','1','2','3','4']  # '정상', 'bearing', 'belt', 'misalignment', 'unbalance'] 
```

```python
current_folder = os.listdir('/daejeon_data/daejeon_data/current(전류)')
kimm_folder = os.listdir('/daejeon_data/daejeon_data/kimm(유선진동)')
vibration_folder = os.listdir('/Desktop/daejeon_data/daejeon_data/vibration(무선진동)')
```

```python
# Data csv 파일 경로 
path_kimm = 'C:/Users/psb31/Desktop/daejeon_data/daejeon_data/kimm(유선진동)/'

path_kimm_oplist = list() # kimm 정상 시계열을 담을 리스트
path_kimm_nolist = list() # kimm 고장 시계열을 담을 리스트

for folder in kimm_folder:
    f1 = path_kimm + folder
    for i in range(1,7):
        f2 = f1 + '/' + str(i)
        f4 = f2 + '/' + '4.op-csv' + '/' +'*.csv'
        path_kimm_oplist = path_kimm_oplist + glob(f4) 
        
        f5 = f2 + '/' + '5.no-csv' + '/' +'*.csv'
        path_kimm_nolist = path_kimm_nolist + glob(f5)
        
```

```python
# 정상시계열 데이터로드

Data_kimm_op = list()
for ts in path_kimm_oplist:
    temp = pd.read_csv(ts, header=0, names=['0','1'], skiprows=3) #, skiprows=3 << 왜 이거 하면 반올림 되는거지...
    Data_kimm_op.append(temp['0'])

Data_kimm_op = np.array(Data_kimm_op)
```

```python
# 정상시계열 라벨
Label_kimm_op = np.zeros(192)
Label_kimm_op
```

```python
# 고장시계열 데이터로드
Data_kimm_no = list()
for ts in path_kimm_nolist:
    temp = pd.read_csv(ts, header=0, names=['0','1'], skiprows=3) #, skiprows=3 << 왜 이거 하면 반올림 되는거지...
    Data_kimm_no.append(temp['0'])

Data_kimm_no = np.array(Data_kimm_no)
Data_kimm_no
```

```python
# 고장시계열 라벨 
# 'bearing', 'belt', 'misalignment', 'unbalance'
kimm_no_label = list()
cnt_bea=0
cnt_bel=0
cnt_mis=0
cnt_unb=0
for name in path_kimm_nolist:
    name = str(name)
    name=name.split('/')
    if name[7][0:3]=='bea':
        kimm_no_label.append('1')
        cnt_bea=cnt_bea+1
        
    elif name[7][0:3]=='bel':
        kimm_no_label.append('2')
        cnt_bel=cnt_bel+1
        
    elif name[7][0:3]=='mis':
        kimm_no_label.append('3')
        cnt_mis=cnt_mis+1
        
    elif name[7][0:3]=='unb':
        kimm_no_label.append('4')
        cnt_unb=cnt_unb+1

kimm_no_label = np.array(kimm_no_label)
kimm_no_label

```

```python
kimm_op = pd.DataFrame(Data_kimm_op)
Label_kimm_op = Label_kimm_op.astype(np.int64)
label_kimm_op = pd.Series(Label_kimm_op, name = 'label')
kimm_op = kimm_op.join(label_kimm_op)
kimm_op
```

```python
kimm_no = pd.DataFrame(Data_kimm_no)
kimm_no_label = kimm_no_label.astype(np.int64)
label_kimm_no = pd.Series(kimm_no_label, name = 'label')
kimm_no = kimm_no.join(label_kimm_no)
kimm_no
```

```python
df = pd.concat([kimm_op,kimm_no],  ignore_index = True)
df.to_csv('kimm.csv')
df
```
